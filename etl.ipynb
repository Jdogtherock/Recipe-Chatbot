#### **Dask Setup**

```python
import dask.dataframe as dd
from dask.distributed import Client
client = Client(memory_limit='4GB')
client
```

#### **Loading Dataset into Dask Dataframe**

```python
raw_recipes = dd.read_csv('data/raw_recipes.csv')
raw_recipes = raw_recipes.repartition(npartitions=25)
```

#### **Data Characteristics**

```python
print(f'SHAPE: {raw_recipes.compute().shape}')
print(f'First 5 Entries:\n {raw_recipes.head(5)}')
```

#### **Removing duplicates that are 100% duplicates (every feature)**

```python
raw_recipes = raw_recipes.drop_duplicates()
print(f'New Shape: {len(raw_recipes)}, {len(raw_recipes[0])}')
```
